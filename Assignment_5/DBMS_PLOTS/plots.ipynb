```python
# CLOCK PLOT FOR SELECT QUERY 

buff_hits = [ 0, 0, 0, 0, 0, 0, 0, 0]
disk_io = [96, 48, 32, 24, 19, 16, 13, 12]

page_sizes = [64, 128, 192, 256, 320, 384, 448, 512]
default_x_ticks = range(len(page_sizes))
plt.figure(figsize=(5, 3))

plt.title('Buffer Hits and Disk I/0 for Varying Page Sizes for Select Query')


plt.xlabel('Page Size (in Bytes)')
plt.ylabel('Number of Buffer Hits or Disk I/0')
plt.plot(page_sizes, disk_io, marker='o', color='red', linestyle='dashed', linewidth=2, markersize=6)
plt.plot(page_sizes, buff_hits, marker='o', color='blue', linestyle='dashed', linewidth=2, markersize=6)
plt.legend(['Disk I/0', 'Buffer Hits'])

plt.xticks(page_sizes, page_sizes)

# Add dots to the plot
plt.show()
```
